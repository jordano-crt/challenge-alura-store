```python
# se agrega una columna que identifica a la tienda
tienda['Tienda'] = 'Tienda 1'
tienda2['Tienda'] = 'Tienda 2'
tienda3['Tienda'] = 'Tienda 3'
tienda4['Tienda'] = 'Tienda 4'

datos_todas_las_tiendas = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)
promedios_envio = datos_todas_las_tiendas.groupby('Tienda')['Costo de envío'].mean()

# grafico
promedios_envio.plot(kind='bar', color='orange')
plt.title('Promedio del costo de envío por tienda')
plt.xlabel('Tiendas')
plt.ylabel('Costo de envío promedio')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
calificaciones_promedio = datos_todas_las_tiendas.groupby('Tienda')['Calificación'].mean()

calificaciones_promedio.plot(kind='line', marker='o', linestyle='-', color='green')
plt.title('Calificación promedio por tienda')
plt.xlabel('Tienda')
plt.ylabel('Calificación promedio')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
facturacion_total = datos_todas_las_tiendas.groupby('Tienda')['Precio'].sum()

for i, valor in enumerate(facturacion_total):
    plt.text(i, valor + 0.02 * valor, f"${valor:,.0f}", ha='center', va='bottom')

facturacion_total.plot(kind='bar', color='dodgerblue')
plt.title('Facturación total por tienda')
plt.xlabel('Tienda')
plt.ylabel('Facturación ($)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

## contexto

El proposito de este challenge es aplicar los conocimientos vistos en los anteriores cursos. Para entrar en contexto, el SR. Juan desea vender una tienda de las 4 que tiene, por que desea iniciar un nuevo emprendimiento.
Entonces se pide determinar cual tienda se debara vender, se aplicaron análisis de sus respectivas cuentas de ventas, para cada tienda. Para así tomar una descisión acertada y no cometer errores. Se tomaron como indicadores de análisis, los ingresos totales de las tiendas, las categorías de productos más y menos vendidas, las calificaciones promedio de los clientes por tienda. los productos más y menos vendidos y el coste de envío promedio para cada tienda.





## Resolución

###Resultados del análisis

**Ingresos totales:**
La tienda con menor ingreso fue la Tienda 4 con  $1,038,375,700. Mientras que la de mayor ingreso fue la Tienda 1, con $1,150,880,400. Esto posiciona a la Tienda 4 como la de menor rendimiento en términos de facturación.

**Calificación promedio de clientes:**
La calificación promedio de la Tienda 4 fue de 4.00, apenas por encima de la Tienda 1 3.98, pero por debajo de la Tienda 2 4.04 y la Tienda 3 4.05, lo que indica una percepción de menor satisfacción por parte de los clientes.

**Categorías más y menos vendidas:**
Indicador considerado como no descisivo para la toma de descisiones.

**Productos más y menos vendidos:**
Al igual que el indicadores de categorias, no considere este factor como relevante para la toma de descisión.

**Costo promedio de envío:**
La Tienda 4 tiene el costo de envío más bajo, con $23,459.46 en promedio, lo que representa una ventaja logística. No obstante, esta ventaja no compensa el bajo ingreso y la calificación moderada.


Tras el análisis de las 4 tiendas, mediante los 5 indicadores clave, se identificó la tienda más adecuada para ser vendida. Si bien no hay mucha diferencia entre los resultados de los indicadores, se consideró a la tienda 4 para la venta, debido a su bajo ingreso total y por ser la segunda tienda menos calificada por los usuarios.

Los indicadores de ventas por categorías y productos más y menos vendidos no generan un impacto significativo para la toma de decisiones, además el indicador de costo promedio del envío es un punto a favor de la tienda 4.
